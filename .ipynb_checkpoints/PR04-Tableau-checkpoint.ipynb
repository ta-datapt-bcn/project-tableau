## Finding the songs from Spotify Charts in the API

```python
unique_songs = songs[['Track', 'Artists']].drop_duplicates().dropna().reset_index(drop=True)
unique_songs['id'] = None
unique_songs['genres'] = ''
unique_songs[keys_list] = None

for idx, item in unique_songs.iterrows():
    try:
        title = item[0] + ' ' + item[1]
        search = sp.search(q = title)
        tracks = search['tracks']['items']
        if tracks == []:
            unique_songs['id'].loc[idx] = None
        elif tracks != []:
            for track in tracks:
                if track['name'].lower() != item[0].lower():
                    pass
                elif track['name'].lower() == item[0].lower():
                    track_id = track['id']
                    unique_songs['id'].loc[idx] = track_id
                    artist_id = sp.track(track_id)['artists'][0]['id']
                    unique_songs['genres'].loc[idx] = sp.artist(artist_id)['genres']
                    features_values = list(sp.audio_features(track_id)[0].values())[:11]
                    unique_songs.loc[idx, keys_list] = features_values
                    print(idx, track['name'], track_id)
                    break
    except:
        print('ERROR', idx, item)
```

```python
unique_songs
```

```python
unique_songs.dropna(how='any', inplace=True)
```

```python
unique_songs.reset_index(inplace=True, drop=True)
```
