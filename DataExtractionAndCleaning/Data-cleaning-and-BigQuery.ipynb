# Cleaning unique_songs_genres.csv

```python
import pandas as pd
import re
```

```python
df = pd.read_csv('unique_songs_genre.csv').drop('Unnamed: 0', axis=1)
```

```python
df['Artists'] = df['Artists'].apply(lambda x: x.split(', '))
```

```python
df['genres'] = df['genres'].apply(lambda x: re.findall('\'(.*?)\'', x))
```

```python
def clean_genres(df, col):
    main_genres = ['latin', 'reggaeton', 'pop', 'trap', 'rap', 'r&b', 'hip hop', 'electro', 'rock', 'edm', 'jazz', 'indie', 'folk', 'house', 'flamenco', 'drill', 'dancehall', 'funk', 'soul', 'urban', 'country', 'rumba']
    for idx, genres in df[col].iteritems():
        for index, genre in enumerate(genres): 
            for main_genre in main_genres:
                if main_genre in genre:
                    df.loc[idx, col][index] = main_genre
                else:
                    pass
```

```python
clean_genres(df, 'genres')
```

```python
df['genres'] = df['genres'].apply(lambda x: list(set(x)))
```

```python
def top3(df, col):
    """
    Input: dataframe, name of a column that contains a list
    Output: 3 new columns with the first 3 items of the list
    """
    new_col_names = [col + '_1', col + '_2', col + '_3']
    df[new_col_names] = None
    for idx, lst in df[col].iteritems():
        try:
            df.loc[idx, new_col_names[0]] = lst[0]
            if len(lst) > 1:
                df.loc[idx, new_col_names[1]] = lst[1]
                if len(lst) > 2:
                    df.loc[idx, new_col_names[2]] = lst[2]
        except:
            continue
```

```python
top3(df, 'Artists')
```

```python
top3(df, 'genres')
```

```python
df.head()
```

```python
df.to_csv('unique_songs_clean.csv', index=False)
```
